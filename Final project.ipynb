```python

total_employees = df.shape[0]

# Calculations
age_28_35_pct = df[(df['age'] >= 28) & (df['age'] <= 35)].shape[0] / total_employees * 100
tenure_3_7_pct = df[(df['length_of_service'] >= 3) & (df['length_of_service'] <= 7)].shape[0] / total_employees * 100
missing_ratings_pct = df['previous_year_rating'].isnull().sum() / total_employees * 100
high_training_kpi_success_pct = df[df['avg_training_score'] > 75]['KPIs_met_more_than_80'].mean() * 100
awards_pct = df['awards_won'].mean() * 100
recruitment_distribution = df['recruitment_channel'].value_counts(normalize=True) * 100

# Build summary DataFrame
eda_summary = pd.DataFrame({
    "Insight": [
        "Age 28–35 %", 
        "Tenure 3–7 years %", 
        "Missing previous year rating %", 
        "KPI success if training score >75 %", 
        "Awards won %"
    ],
    "Value": [
        round(age_28_35_pct, 2), 
        round(tenure_3_7_pct, 2), 
        round(missing_ratings_pct, 2), 
        round(high_training_kpi_success_pct, 2), 
        round(awards_pct, 2)
    ]
})

# Add recruitment distribution separately
recruitment_df = pd.DataFrame({
    "Insight": [f"Recruitment channel: {k}" for k in recruitment_distribution.index],
    "Value": recruitment_distribution.round(2).values
})

# Combine into one summary table
eda_summary = pd.concat([eda_summary, recruitment_df], ignore_index=True)

print(eda_summary)

# Optional: save to Excel
# eda_summary.to_excel("eda_summary.xlsx", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- CALCULATIONS ---

# 1. Age 28–35 %
age_28_35_pct = df[(df['age'] >= 28) & (df['age'] <= 35)].shape[0] / total_employees * 100

# 2. Tenure 3–7 years %
tenure_3_7_pct = df[(df['length_of_service'] >= 3) & (df['length_of_service'] <= 7)].shape[0] / total_employees * 100

# 3. Missing previous year rating %
missing_ratings_pct = df['previous_year_rating'].isnull().sum() / total_employees * 100

# 4. KPI success for training score > 75 %
high_training_kpi_success_pct = df[df['avg_training_score'] > 75]['KPIs_met_more_than_80'].mean() * 100

# 5. Awards %
awards_pct = df['awards_won'].mean() * 100

# 6. Recruitment channel distribution
recruitment_distribution = df['recruitment_channel'].value_counts(normalize=True) * 100

# --- SUMMARY TABLE ---
eda_summary = pd.DataFrame({
    "Insight": [
        "Age 28–35 %",
        "Tenure 3–7 years %",
        "Missing previous year rating %",
        "KPI success if training score >75 %",
        "Awards won %"
    ],
    "Value": [
        round(age_28_35_pct, 2),
        round(tenure_3_7_pct, 2),
        round(missing_ratings_pct, 2),
        round(high_training_kpi_success_pct, 2),
        round(awards_pct, 2)
    ]
})

recruitment_df = pd.DataFrame({
    "Insight": [f"Recruitment channel: {k}" for k in recruitment_distribution.index],
    "Value": recruitment_distribution.round(2).values
})

eda_summary = pd.concat([eda_summary, recruitment_df], ignore_index=True)
print(eda_summary)

# --- VISUALIZATIONS ---

# 1. Age distribution histogram
plt.figure(figsize=(6,4))
df['age'].hist(bins=20, edgecolor='black')
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# 2. Length of service histogram
plt.figure(figsize=(6,4))
df['length_of_service'].hist(bins=20, edgecolor='black')
plt.title("Length of Service Distribution")
plt.xlabel("Years of Service")
plt.ylabel("Count")
plt.show()

# 3. Recruitment channel pie chart
plt.figure(figsize=(6,6))
df['recruitment_channel'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Recruitment Channel Distribution")
plt.ylabel("")  # remove y-label
plt.show()

# 4. KPI vs Training Score (boxplot)
plt.figure(figsize=(6,4))
df.boxplot(column='avg_training_score', by='KPIs_met_more_than_80')
plt.title("Training Score by KPI Achievement")
plt.suptitle("")  # remove default pandas title
plt.xlabel("KPI met > 80% (0=No, 1=Yes)")
plt.ylabel("Average Training Score")
plt.show()

# 5. Awards vs KPI (bar chart)
awards_kpi = df.groupby('awards_won')['KPIs_met_more_than_80'].mean() * 100
awards_kpi.plot(kind='bar', figsize=(6,4))
plt.title("KPI Success Rate by Awards Won")
plt.xlabel("Awards Won (0=No, 1=Yes)")
plt.ylabel("KPI Success %")
plt.show()
```

```python
# Ensure column is numeric
import numpy as np
df["no_of_trainings"] = pd.to_numeric(df["no_of_trainings"], errors="coerce")

# Group, calculate KPI %, and sort by number of trainings
training_count_vs_kpi = (
    df.groupby("no_of_trainings")["KPIs_met_more_than_80"]
    .mean()
    .mul(100)
    .sort_index()
)

# Reindex explicitly for 1–9 order (fill missing with 0)
training_count_vs_kpi = training_count_vs_kpi.reindex(np.arange(1, 10), fill_value=0)

# Plot
plt.figure(figsize=(7,4))
bars = plt.bar(
    training_count_vs_kpi.index,
    training_count_vs_kpi.values,
    color="#c6e374",
    edgecolor="black"
)

# Add % labels
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 1,
        f"{height:.1f}%",
        ha="center",
        va="bottom",
        fontsize=10
    )

# 
plt.ylabel("Average KPIs Rate")
plt.xlabel("Number of Trainings", fontsize=11)
plt.xticks(range(1, 10))
plt.ylim(0, training_count_vs_kpi.max() + 10)

# Remove y-axis & grid for cleaner look
plt.yticks([])
plt.gca().spines["left"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["top"].set_visible(False)
plt.grid(False)

plt.tight_layout()
plt.show()
```
